```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/dichika/mydata/master/ore.csv")
data["time"] = pd.to_datetime(data["time"])
```

```python
data
```

```python
data["weekday"] = [dt.weekday() for dt in data["time"]]
data
```

月曜0～日曜6

```python
data.groupby("weekday").mean()
```

```python
print(pd.rolling_mean(data["steps"],2))
print(pd.rolling_mean(data["steps"],2,1))
print(pd.rolling_mean(data["steps"],7))
print(pd.ewma(data["steps"],span=2))
```

pd.rolling_mean : 単純移動平均
pd.ewma : 指数移動平均

```python
plt.subplot(211)
plt.plot(data["time"],data["steps"])
plt.subplot(212)
plt.plot(data["time"],pd.rolling_mean(data["steps"],7,1))
plt.ylim(0,30000)
```
